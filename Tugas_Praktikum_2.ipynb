Membuat double linked list dengan 3 node di dalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci.

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None

n1 = Node("Magelang")
n2 = Node("Yogyakarta")
n3 = Node("Bandung")

class LinkedList:
    def __init__(self):
        self.headval = None
        self.tailval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval, end="\n" if printval.nextval else "\n")
            printval = printval.nextval

    def AddInbetween(self, mid_node, newdata):
        if mid_node is None:
            print("The mentioned node is absent")
            return

        NewNode = Node(newdata)

        NewNode.nextval = mid_node.nextval
        NewNode.prevval = mid_node

        if mid_node.nextval is not None:
            mid_node.nextval.prevval = NewNode
        else:
            self.tailval = NewNode

        mid_node.nextval = NewNode

    def RemoveNode(self, Removekey):
        CurrentNode = self.headval

        while CurrentNode is not None:
            if CurrentNode.dataval == Removekey:
                if CurrentNode.prevval is not None:
                    CurrentNode.prevval.nextval = CurrentNode.nextval
                else:
                    self.headval = CurrentNode.nextval

                if CurrentNode.nextval is not None:
                    CurrentNode.nextval.prevval = CurrentNode.prevval
                else:
                    self.tailval = CurrentNode.prevval

                print(f"Node dengan data '{Removekey}' berhasil dihapus")
                return

            CurrentNode = CurrentNode.nextval

        print(f"Node dengan data '{Removekey}' tidak ditemukan")

Li = LinkedList()

n1.nextval = n2
n2.prevval = n1
n2.nextval = n3
n3.prevval = n2

Li.headval = n1
Li.tailval = n3

print("=== Double Linked List Awal ===")
Li.listprint()

print("\n=== Menyisipkan 'Semarang' setelah 'Magelang' ===")
Li.AddInbetween(n1, "Semarang")
Li.listprint()

print("\n=== Setelah menghapus node 'Magelang' ===")
Li.RemoveNode("Magelang")
Li.listprint()
```

Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan
akhir dan hapus akhir !

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head  #
        else:

            current = self.head
            while current.next != self.head:
                current = current.next

            current.next = new_node

            new_node.next = self.head

    def pop(self):
        if self.head is None:
            print("List kosong, tidak ada yang dihapus")
            return None

        if self.head.next == self.head:
            removed_data = self.head.data
            self.head = None
            return removed_data

        current = self.head
        while current.next.next != self.head:
            current = current.next

        removed_node = current.next
        removed_data = removed_node.data

        current.next = self.head
        return removed_data

    def display(self, putaran=1):
        if self.head is None:
            print("List kosong")
            return
        nodes = []
        current = self.head

        for _ in range(putaran * self.__len__()):
            nodes.append(str(current.data))
            current = current.next
        print("\n".join(nodes) + " (circular)")


    def __len__(self):
        if self.head is None:
            return 0
        count = 1
        current = self.head
        while current.next != self.head:
            count += 1
            current = current.next
        return count

cll = CircularLinkedList()
cll.append("Magelang")
cll.append("Yogyakarta")
cll.append("Bandung")

print("=== Circular Linked List awal (3 node) ===")
cll.display(putaran=2)

cll.append("Semarang")
print("\n=== Setelah penyisipan akhir (tambah 'Semarang') ===")
cll.display(putaran=2)

removed = cll.pop()
print(f"\n=== Setelah penghapusan akhir (menghapus '{removed}') ===")
cll.display(putaran=2)
```
